```python
#import libraries
import pandas as pd
import numpy as np
import tensorflow as tf
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load the dataset
from tensorflow.keras.datasets import cifar100
```

```python
# Load training and testing data
(X_train, y_train), (X_test, y_test) = cifar100.load_data(label_mode='fine')
```

```python
# Print the shape of the data
print("Training data shape:", X_train.shape)
print("Testing data shape:", X_test.shape)
print("Training labels shape:", y_train.shape)
print("Testing labels shape:", y_test.shape)
```

```python
# Dataset characteristics
print(f"Training data shape: {X_train.shape}")
print(f"Test data shape: {X_test.shape}")
print(f"Number of classes: {len(np.unique(y_train))}")
```

```python
# Display basic statistics of the dataset
print(f"Image dimensions: {X_train.shape[1:]} (Height x Width x Channels)")
print(f"Number of training samples: {X_train.shape[0]}")
print(f"Number of test samples: {X_test.shape[0]}")
```

```python
# Check distribution of labels in the training set
unique, counts = np.unique(y_train, return_counts=True)
print("Class distribution in training data:")
print(dict(zip(unique, counts)))
```

```python
#Training a logistic regression model on the CIFAR100 data
```

```python
X_train_flat = X_train.reshape(X_train.shape[0], -1)
X_test_flat = X_test.reshape(X_test.shape[0], -1)
```

```python
#normalization
```

```python
# Normalize pixel values to [0, 1]
X_train = X_train.astype("float32") / 255.0
X_test = X_test.astype("float32") / 255.0
```

```python
#scale the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_flat = scaler.fit_transform(X_train_flat)
X_test_flat = scaler.transform(X_test_flat)
```

```python
#train the logistic regression model
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(multi_class='multinomial', solver='lbfgs', max_iter=1000)
clf.fit(X_train_flat, y_train)
```

```python
print("Training Accuracy:", clf.score(X_train_flat, y_train))
print("Test Accuracy:", clf.score(X_test_flat, y_test))
```

```python
#the training accuracy of the Logistic model is 72.442%
#the test accuracy of the Logistic model is 12%
```

```python
preds = clf.predict(X_test_flat)
```

```python
from sklearn.metrics import confusion_matrix

print(confusion_matrix(y_test,preds))
```

```python
cm = confusion_matrix(y_test, preds, labels=clf.classes_)

# Step 4: Plot Confusion Matrix
plt.figure(figsize=(100, 100))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=clf.classes_, yticklabels=clf.classes_)
plt.title("Confusion Matrix", fontsize=16)
plt.xlabel("Predicted Labels", fontsize=14)
plt.ylabel("True Labels", fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12, rotation=0)
plt.show()
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,preds))
```
